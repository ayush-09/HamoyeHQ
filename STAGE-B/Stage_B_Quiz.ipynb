```python
# Training set of the T2 and T6 columns 
train_x = X_train[['T2']]
train_y = X_train[['T6']]

# Testing set of the T2 and T6 column 
test_x = X_test[['T2']]
test_y = X_test[['T6']]
```

```python
model = LinearRegression()
model.fit(train_x, train_y)

p = model.predict(test_x)
```

```python
##Linear Regression
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred=lr.predict(X_test)


accuracy_matrix=pd.DataFrame(data= [round(mean_absolute_error(y_test, y_pred),2),
                        round(mean_squared_error(y_test, y_pred),2),
                       round(np.sum(np.square(y_test- y_pred)),2), 
                       round(r2_score(y_test, y_pred),2),
                       round(np.sqrt(mean_squared_error(y_test, y_pred)),3)], 
                 index= ["Mean Absolute Error",
                        "Mean Square Error",
                         "Residual Error(RSS)",
                         "R2 Score (CoD)", "Root mean Square"],
                 columns= ['Linear Regression'])
```

```python
##weights

def get_weights(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df
```

```python
linear_model_weights= get_weights(lr, X_train, "Linear_Model_Weight" )
```
